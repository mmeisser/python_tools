## import all data from csv files

```
import pandas as pd
import io

dataframe_collection = {}

listed = !ls "/content/drive/My Drive/logs"
for index, file in enumerate(listed):
  item = '/content/drive/My Drive/logs/' + (file)
  if "csv" not in file:
    print("skipping  ... " + item)
  else:
    print("importing [" + str(index) + "] ... " + item)
    df = pd.read_csv(item)
    #print(df)
    dataframe_collection[index] = df
```

## plot data

```
import plotly.graph_objs as go
import csv
import matplotlib.pyplot as plt
import plotly.express as px
import orca

import os

root_dir = "/content/drive/My Drive/logs"
results_folder = "/results/"

# check if your project folder exists. if not, it will be created.
if os.path.isdir(root_dir + results_folder) == False:
  os.mkdir(root_dir + results_folder)
  print(root_dir + results_folder + ' did not exist but was created.')

# change the OS to use your project folder as the working directory
os.chdir(root_dir + results_folder)

for index in dataframe_collection:
 df = dataframe_collection[index]
 if ' Device type' in df:
  print("detected akra file format")
  is_primary = df[' Device type']== ' Primary' #this is a column with true/false to assert if this is data for a controller
  stylus_data = df[is_primary]    #therefore we only take that subset of lines for the controller (ie stylus)
  time=stylus_data['TimeStamp (ms)']
 else:
  print("detected sartoinput file format")
  stylus_data = df
  time=stylus_data['TimeStamp (sec)']

 nib=stylus_data[' Touchpad X value']*32768
 primary=stylus_data[' Touchpad Y value']*32768


# plot graph with all series
 fig = go.Figure([
    go.Scatter(
        name='Nib',
        y=nib,
        mode='lines',
        marker=dict(color='blue', size=2),
        showlegend=True
    ),
    go.Scatter(
        name='Primary',
        y=primary,
        mode='lines',
        marker=dict(color='red', size=2),
        showlegend=True
    )
])
 fig.update_layout(
    yaxis_title='adc raw value [counts]',
    title=listed[index],
    hovermode="x"
)
 fig.show()
 fig.write_image(listed[index] +".png")
```
